### REMOVING OUTLIERS

```python
from scipy.stats import zscore
z=np.abs(zscore(df))
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
# rows and column
z[27][6]
```

```python
z[40][1]
```

```python
z[64][1]
```

```python
df_new=df[(z<3).all(axis=1)]
print("before removing the outliers:",df.shape)
print("after removing the outliers:",df_new.shape)
```

### USING IQR

```python
from scipy import stats 
IQR = stats.iqr(df) 
IQR
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
```

```python
IQR = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(IQR.shape)
```

Using zscore method we can notice there are 9 rows have been removed.

Using IQR method we can see it is removing 10 row 


so considering the ZSCORE.

```python
from sklearn.model_selection import cross_val_score

np.random.seed(10)
def rmse_cv(model, x,y):
    rmse =- (cross_val_score(model, x,y, scoring='neg_mean_squared_error', cv=10))
    return(rmse)


models = [LinearRegression(),
             SVR(kernel='linear'),
             SVR(kernel='poly'),
             SVR(kernel='rbf'),
             RandomForestRegressor(),
             DecisionTreeRegressor(),  
             ]
            


names = ['LR','svr','svr_p','svr_r','RF','DTR']

for model,name in zip(models,names):
    score = rmse_cv(model,x,y)
    print("{}    : {:.6f}, {:4f}".format(name,score.mean(),score.std()))
```
