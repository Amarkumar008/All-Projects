### Using IQR

```python
from scipy import stats 
IQR = stats.iqr(df) 
IQR
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
```

```python
IQR = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(IQR.shape)
```

Using zscore we can see there are 1556 no rows have been removed.

Using IQR we can notice 756 no of rows have been remove.

there is huge data loss in "zscore" so considering "IQR" method.

```python
df=IQR
```

```python
df.shape
```

# HYPER PARAMETER TUNING:

```python
from sklearn.model_selection import GridSearchCV
```

```python
model=LogisticRegression()
param_grid={
    'max_iter' : [100],
    'solver':['liblinear','newton-cg'],
     'random_state':[10]              
}
```

```python
model_grid=GridSearchCV(LogisticRegression(),param_grid,cv=4,scoring='accuracy',n_jobs=-1,verbose=2)
```
